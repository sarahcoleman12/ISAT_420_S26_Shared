# Download Script for USGS station data

USGS Station data is available through and API. This script makes use of the API to download the data for the South River Station.
We will talk about APIs in general soon. 

```python
# this is a package that you won't have installed
# we will talk about this in class
from dataretrieval import nwis
```

```python
# Set the parameters needed to retrieve data
siteNumber = "01626000" # South River Near Waynesboro, VA - USGS-01626000
parameterCode = "00060" # Discharge
startDate = "1952-10-01"

# Retrieve the data
dailyStreamflow = nwis.get_dv(sites=siteNumber, parameterCd=parameterCode, start=startDate)
dailyStreamflow
```

```python
import pandas as pd
file_out = f'../Data/usgs_nwis_{siteNumber}_{startDate}.csv' 
dailyStreamflow[0].to_csv(file_out)
```
